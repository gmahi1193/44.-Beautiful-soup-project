### Downloading images from site

```python
# import libraries

import bs4
import requests
import cv2
import matplotlib.pyplot as plt
```

```python
# considering url for image downloads

url = "https://www.passiton.com/inspirational-quotes?page=2"
```

```python
# getting content from url

response = requests.get(url)
```

```python
# see downloaded content

response.content
```

```python
# html parser

soup = bs4.BeautifulSoup(response.content)

# after inspecting site, finding content for site
ima_element = soup.findAll('img')
ima_element = ima_element[2:]
ima_element
```

```python
# iterating content for download

for i in range (10):
    with open ('inspiration{}.jpg'.format(i), 'wb') as file:
        img_url = ima_element[i].attrs['src']
        response = requests.get(img_url)
        file.write(response.content)
        
```

```python
# Checking the downloaded images

for i in range(10):
    plt.imshow(cv2.cvtColor(cv2.imread('inspiration{}.jpg'.format(i)), cv2.COLOR_BGR2RGB))
    plt.axis('off')
    plt.show()
```
